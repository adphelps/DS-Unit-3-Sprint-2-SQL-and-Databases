```python
# - (Bonus! Hard, may require pulling and processing with Python) How many married
#   couples were aboard the Titanic? Assume that two people (one `Mr.` and one
#   `Mrs.`) with the same last name and with at least 1 sibling/spouse aboard are
#   a married couple.
query = """
SELECT "Titanic"."Name" as n
FROM "public"."Titanic"
WHERE "Titanic"."Siblings/Spouses Aboard" > 0
"""
pg_cur.execute(query)
names = pg_cur.fetchall()
names
```

```python
all_names = [x[0] for x in names]
names_we_care_about = [(x.split()[0], x.split()[-1]) for x in all_names if x.split()[0] in ['Mr.', 'Mrs.']]
found_names = {}
married_count = 0
for prefix, name in names_we_care_about:
    if name in found_names:
        
        if found_names[name] != prefix:
            married_count += 1
    else:
        found_names[name] = prefix
print("number of married couples: ", married_count)
```
